```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_fes_2d(data1, data2, N, bin_edges, weight=None):
    if weight is None:
        weight = np.ones(len(data1[:N]))

    hist, xedges, yedges = np.histogram2d(
        data1[:N], data2[:N],
        bins=bin_edges,
        density=True,
        weights=np.exp(weight[:N] / (kB * T))
    )

    fes = -np.log(hist + 1e-10) * kB * T
    fes -= np.nanmin(fes)

    xcenters = 0.5 * (xedges[:-1] + xedges[1:])
    ycenters = 0.5 * (yedges[:-1] + yedges[1:])

    return fes.T, xcenters, ycenters
```

```python
plt.figure(figsize=(9,3))
data_all = data = np.loadtxt(
            f"/ocean/projects/mth210003p/lyuliyao/meta_bias/chignolin/1VII2/"
            f"case0.5/run1/COLVAR"
        )
ax1 = plt.subplot(1,3,1)   # 上面
ax2 = plt.subplot(1,3,2)   # 下面
ax3 = plt.subplot(1,3,3)   # 下面
for run_id in range(1, 50):
    try:
        data = np.loadtxt(
            f"/ocean/projects/mth210003p/lyuliyao/meta_bias/chignolin/1VII2/"
            f"case0.5/run{run_id}/COLVAR"
        )
        if run_id >1 :
            data_all = np.concatenate([data_all,data],axis=0)
        if run_id%3 ==0:
            ax1.hist(data_all[:,-5], bins=100, density=True, histtype='step', label=f"run{run_id}")
            ax2.hist(data_all[:,-4], bins=100, density=True, histtype='step', label=f"run{run_id}")
            ax3.hist(data_all[:,-3], bins=100, density=True, histtype='step', label=f"run{run_id}")
        plotted = True
    except Exception:
        continue

if plotted:
    plt.legend()
plt.show()


kB = 8.314462618e-3
T = 300.0
# fes,x,y = compute_fes_2d(data_all[:,2],data_all[:,3],N=10000000,bin_edges=100,weight=data_all[:,-1])
# plt.contourf(x,y,fes)
# plt.show()
fes,x,y = compute_fes_2d(data_all[:,1],data_all[:,3],N=10000000,bin_edges=100)
plt.contourf(x,y,fes)
plt.show()
# fes,x,y = compute_fes_2d(data_all[:,1],data_all[:,2],N=10000000,bin_edges=100,weight=data_all[:,-1])
# plt.contourf(x,y,fes)
# plt.show()
plt.plot(data_all[::10,3],".")
plt.plot(data_all[::10,1],".")
```
